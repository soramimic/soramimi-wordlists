```python
import os
import pandas as pd
PATH = os.path.join("data","baseball_players2022diff.tsv")
#PATH = os.path.join("data","baseball_players_sample.tsv")

names = ["name","team","dummy0","full_pron","family_pron","given_pron"]
df = pd.read_csv(PATH, sep="\t",header=None, names = names)
# dummy0列を削除する
df = df.drop(columns=["dummy0"])
df.head()
```

```python
#nameをカッコでバラす関数を定義
import re
def format_name(name):
  """_summary_

  Args:
      name (str): name

  Returns:
      list[str]: names
  """
  # 全角カッコを半確認する
  name = name.translate(str.maketrans("（）", "()"))
  #スペース削除
  name = "".join(name.split())
  #カッコでスプリット
  namelist = re.split("[()]",name)
  # 空の要素を削除
  namelist = [x for x in namelist if x != ""]
  # あいうえおの昇順ソート
  namelist.sort()
  return namelist

print(format_name("奥田和男（旭和男）"))
print(format_name("旭和男（奥田和男）"))
```

```python
def get_ids(names, teams, cnt={}):  
  ids = []
  for name, team in zip(names, teams):
    name = format_name(name)[0]
    team = team.split("-")[0]
    id_head = "{}_{}".format(name, team)
    id = "{}_{}".format(id_head, str(cnt.get(id_head,0)).zfill(5))
    ids.append(id)
    cnt[id_head] = cnt.get(id_head,0) + 1
  return ids

get_ids(df["name"],df["team"],{})

  
```

```python
df["id"] = get_ids(df["name"],df["team"])
df
```

```python
# 姓名分離するためのクラス
from namedivider import NameDivider
class Divider:
  def __init__(self):
    self.name_divider = NameDivider()
  def devide_jp_name(self,name):
    divided_name = self.name_divider.divide_name(name).to_dict()
    return divided_name
  def devide_foreign_name(self,name):
    sep = '・'
    family = name.split(sep)[-1]
    given = sep.join(name.split(sep)[:-1])
    #sepが複数含まれていたらscoreを下げる
    score = 1/(len(name.split(sep))-1)
    return {'family':family, 'given':given, 'score':score, 'algorithm': 'foreign_rule'}
  def divide_name(self,name):
    sep = '・'
    if len(name)==1:
      return {'family':name, 'given':'', 'score':1, 'algorithm': 'one_char_rule'}
    elif sep in name:
      return self.devide_foreign_name(name)
    else:
      return self.devide_jp_name(name)

divider = Divider()
df["name"].apply(divider.divide_name) 
```

```python
# 名前候補を列方向に展開
def expand_name(df_rows,name_key="name"):
  new_rows = []
  for index,row in df_rows.iterrows():
    name = row[name_key]
    namelist = format_name(name)
    for n in namelist:
      new_row = row.copy()
      new_row[name_key] = n
      new_rows.append(new_row)
  return pd.DataFrame(new_rows)
```

```python

def add_separated_name_column(df, name_key="name"):
  df = df.copy()

  new_rows = []
  for index, row in df.iterrows():
    name = row[name_key]
    separated_name = divider.divide_name(name)
    row = {k:v for k,v in row.items()}
    row.update(separated_name)
    new_rows.append(row)
  
  return pd.DataFrame(new_rows)

add_separated_name_column(expand_name(df))
```

```python
# フルネーム、姓、名を別行に分けて、列名を選手表の形式にする
def make_player_table(df):
  new_rows = []
  for index, row in df.iterrows():
    surfaces = [row["name"],row["family"],row["given"]]
    pronunciations = [row["full_pron"],row["family_pron"],row["given_pron"]]
    types = ["full","family","given"]
    team = row["team"]
    id = row["id"]
    original = row["original"]
    for surface, pronunciation, type in zip(surfaces, pronunciations, types):
      if not surface: continue
      new_row = {
        "id":id,
        "original":original,
        "team":team,
        "surface": surface,
        "pronunciation": pronunciation,
        "type": type,
        "org_id":id,
      }
      new_rows.append(new_row)
  return pd.DataFrame(new_rows)

df2 = df.copy()
df2["original"] = df["name"]
df2 = expand_name(df2)
df2 = add_separated_name_column(df2)
df2 = make_player_table(df2)
df2

```
