```python
import pandas as pd
PATH = "../words_tidy/baseball.csv"
df = pd.read_csv(PATH)
df.head()
```

```python
def df2json(df):
  obj = {}
  for index, row in df.iterrows():
    surface, original = row["surface"], row["original"]
    obj[surface] = obj.get(surface, []) + [original]
  return obj

import json
json.dump(df2json(df), open("../words_tidy/baseball.json", "w"), indent=2,ensure_ascii=False)
```
